# Data Visualization

## Univariate Analysis

```python
# categorical columns

for i in cat_cols:
    print(i)
    sns.countplot(df[i])
    plt.title(i)
    plt.xticks(rotation=90)
    plt.show()
```

**Inferences**:

1. White women has more breast cancer then black and others.
2. Married women tends to have breast cancer than single, divorced or widowed.
3. Moderately differentiate people have breast cancer than poorand well differentiated.
4. Grade 2 people have breast cancer.
5. Regional people have braest cancer.
6. People who have positive estrogen status have breast cancer.
7. People who have positive progesterone status have breast cancer.

## Bivariate Analysis

```python
# bi-varaiate analysis  of cat_cols with the target columns which is "Status"


plt.figure(figsize=(20,12))

plt.subplot(3,2,1)
sns.countplot(df['Race'],hue=df.Status,)
plt.show()


plt.figure(figsize=(20,12))
plt.subplot(3,2,2)
sns.countplot(df['Marital Status'],hue=df.Status)
plt.show()


plt.figure(figsize=(20,12))
plt.subplot(3,2,3)
sns.countplot(df['differentiate'],hue=df.Status)
plt.show()


plt.figure(figsize=(20,12))
plt.subplot(3,2,4)
sns.countplot(df['Grade'],hue=df.Status)
plt.show()


plt.figure(figsize=(20,12))
plt.subplot(3,2,5)
sns.countplot(df['Estrogen Status'],hue=df.Status)
plt.show()


plt.figure(figsize=(20,12))
plt.subplot(3,2,6)
sns.countplot(df['Progesterone Status'],hue=df.Status)
plt.show()

```

**Inferences**:

We can see that from the above graphs that people who have breast cancer  and treated are still alive.

## Multivariate Analysis

```python
plt.figure(figsize=(45,32))
sns.catplot(x= "Grade",y= "Tumor Size",kind="bar",hue='Status',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(45,32))
sns.catplot(x= "Marital Status",y= "Tumor Size",kind="violin",hue='Status',data=df,split=True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(45,32))
sns.catplot(x= "Race",y= "Tumor Size",kind="violin",hue='Status',data=df,split=True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(45,32))
sns.catplot(x= "Estrogen Status",y= "Tumor Size",kind="bar",hue='Status',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(45,32))
sns.catplot(x= "Progesterone Status",y= "Tumor Size",kind="bar",hue='Status',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(45,32))
sns.catplot(x= "A Stage",y= "Survival Months",kind="bar",hue='Status',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x= "Reginol Node Positive",y= "Regional Node Examined",hue='Status', data=df)
plt.xticks(rotation=90)
plt.show()
```

**Inferences:**
    
- If women diagnosis with breast cancer and the stage is **initial** then they have higher chances of living.


- Most people who has regional and distant stages are still alive, so its all about the treatment and the age.
